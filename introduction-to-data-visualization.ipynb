```python
# Gerekli olan kütüphaneleri yükleyelim
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Seattle ve Austin veri setleri ile başlayalım
* Seattle: Amerika Birleşik Devletleri'nin Washington eyaletinde bulunan bir şehir.
* Austin: Amerika Birleşik Devletleri'nin Teksas eyaletinde bulunan bir şehir.
* MLY-PRCP-NORMAL: ortalama yağış için kullanacağız
* MLY-TAVG-NORMAL: ortalama sıcaklık için kullanacağız

```python
# "STATION" sütun değerleri "USW00094290" değerine eşit olanları filtreleyip seattle_weather adındaki yeni değişkene aktaralım.
seattle_weather = weather[weather["STATION"] == "USW00094290"] 
```

## Zaman serisini çizme
* Verileri bir zaman indeksi ile okuyalım
* parse_dates: Bu anahtar sözcük ile varsayılan dizeleri, float sayıları veya tam sayıları tarihlere dönüştürebiliyoruz.
* index_col: Bu anahtar sözcüğe indeks değerlerini içeren sütunun numarasını gireceğiz. read_csv() ile okutursak ve satırlar 1. sutündaysa index_col=0 deriz read_csv içine.

```python
import pandas as pd

# read_csv kullanarak dosyadaki verileri okuyalım
climate_change = pd.read_csv('datasets/climate_change.csv', parse_dates=["date"], index_col="date")
```

```python
climate_change.reset_index().info()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Çizgi grafiğine "relative_temp" için zaman serisi ekleyelim
# x eksenine zaman dilimi yani yıllar, y eksenine ise bağıl sıcaklık değerlerini girelim
plt.plot( climate_change['relative_temp'], label="Bağıl Sıcaklık")

plt.title("Yıllara Göre Bağıl Sıcaklık Miktarları")
plt.xlabel('Yıllar')
plt.ylabel('Bağıl Sıcaklık (Santigrat)')
plt.legend()
plt.show()
```

## *Zaman serisi verilerini çizen bir fonksiyon tanımlayalım*

```python
# plot_timeseries adlı bir fonksiyon tanımlayalım
def plot_timeseries(axes, x, y, color, xlabel, ylabel):

  # x,y girişlerini sağlanan renkte çizin
  axes.plot(x, y, color=color)
  axes.set_xlabel(xlabel) # x eksenini isimlendirelim
  axes.set_ylabel(ylabel, color=color) # y eksenini isimlendirelim

  # y ekseni için renk parametreleri ayarlayalım
  axes.tick_params('y', colors=color)
```

### *Tanımladığımız fonksiyonu kullanalım*

```python
fig, ax = plt.subplots()

# CO2 seviyeleri zaman serisini mavi ile gösterelim
plot_timeseries(ax, climate_change.index, climate_change['co2'], "blue", 'Yıllar', 'CO2 Seviyeleri')

# x eksenini paylaşan ikiz bir Eksen nesnesi oluşturalım ve bunu ax2 adındaki değişkene aktaralım
ax2 = ax.twinx()

# Bağıl sıcaklık verilerini kırmızı ile gösterelim
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], "red", 'Yıllar', 'Bağıl Sıcaklıklar (santigrat)')

plt.show()
```

### *Bir zaman serisi grafiğine açıklama ekleme*

```python
fig, ax = plt.subplots()

# Bağıl sıcaklık verilerini çizelim
ax.plot(climate_change.index, climate_change["relative_temp"], color='red')

# Sıcaklıkların 1 dereceyi aştığı tarihi not edelim
ax.annotate(">1 derece", xy=(pd.Timestamp('2015-10-06'), 1))
#tarih bilgisini Timestamp() ile alıyorum, annotate() ile 1 dereceyi geçen bölgeyi işaretliyorum

plt.show()
```

### *Zaman serisini çizmek: hepsini bir araya getirmek*

```python
fig, ax = plt.subplots()

# CO2 seviyeleri zaman serisini mavi ile gösterelim
plot_timeseries(ax, climate_change.index, climate_change['co2'], 'darkviolet', 'Yıllar', 'CO2 Seviyeleri')

# x eksenini paylaşan bir Eksen nesnesi oluşturalım ve bunu ax2 adındaki değişkene atalım
ax2 = ax.twinx()

# Bağıl sıcaklık verilerini kırmızı ile gösterelim
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], 'y', 'Yıllar', 'Bağıl Sıcaklık (Santigrat)')

# Bağıl sıcaklık değeri 1 dereceden büyük olan noktayı gösterelim
ax2.annotate(">1 derece", xy=(pd.Timestamp('2015-10-06'), 1), xytext=(pd.Timestamp('2008-10-06'), -0.2), arrowprops={"arrowstyle":"->", "color": "red"})

plt.show()
```

## *Summer 2016 veri seti ile devam edelim*
- *mens_rowing: Erkek kürek sporcusu*
- *mens_gymnastics: Erkek jimnastik sporcusu*

```python
fig, ax = plt.subplots()

#Kürek sporu (Rowing) olanları filtreleyelim ve bunu mens_rowing adlı değişkene aktaralım
mens_rowing = summer_2016[(summer_2016["Sport"] == "Rowing") & (summer_2016["Sex"] == "M")]

# Jimnastik (Gymnastics) sporu olanları filtreleyelim ve bunu mens_gymnastics adlı değişkene aktaralım
mens_gymnastics = summer_2016[(summer_2016["Sport"] == "Gymnastics") & (summer_2016["Sex"] == "M")]

# mens_rowing için bir "Ağırlık" histogramı çizelim
plt.hist(mens_rowing["Weight"], label="Erkek Kürek Sporcusu")

# mens_gymnastics için bir "Ağırlık" histogramı çizelim
plt.hist(mens_gymnastics["Weight"], label="Erkek Jimnastik Sporcusu")

plt.xlabel("Ağırlıklar (kg)")
plt.ylabel("Yaşlar")
plt.legend()
plt.show()
```

### *Bazı Önemli Anahtar Kelimeleri Tanıyalım*
- *histtype: Histogramın tipini belirlemenizi sağlar. örn: bar', 'barstacked', 'step', 'stepfilled'*
- *bins: Histogramdaki kutu sayısı diyebiliriz."auto" değer vererek otomatik olarak da görebilirsiniz. bins=[9,18,25] bu şekilde de değer alabilir. Bunun anlamı 9'dan 18'e kadar böl,18'den 25'e kadar böl.*
- *rwidth: Çubukların genişliğini belirler. 0-1 arasında değer alır.*
- *orientation: Grafiğin yatay ya da dikey çizilmesini belirler. "horizontal" ve "vertical değerler alır. Horizontal yatay, vertical dikey demektir.*

```python
fig, ax = plt.subplots()

# mens_rowing için bir "Ağırlık" histogramı çizelim
ax.hist(mens_rowing["Weight"], label="Rowing", bins=5, histtype="step", color="royalblue", orientation="horizontal")

# mens_gymnastics için bir "Ağırlık" histogramı çizelim
ax.hist(mens_gymnastics["Weight"], label="Gymnastics", bins=5, histtype="step", color="red", orientation="horizontal")

plt.xlabel("Ağırlıklar (kg)")
plt.ylabel("Yaşlar")
plt.legend()
plt.show()
```

### *Bir çubuk grafiğe hata çubukları ekleme*

```python
fig, ax = plt.subplots()

# Add a bar for the rowing "Height" column mean/std
# Erkek kürek sporcuların boy uzunluklarının ortalamasını alalım ve boy uzunluklarının standart sapmasını kullanarak hata çubuklarını çizelim
plt.bar("Rowing", mens_rowing["Height"].mean(), yerr=mens_rowing["Height"].std())

# Erkek jimnastik sporcuların boy uzunluklarının ortalamasını alalım ve boy uzunluklarının standart sapmasını kullanarak hata çubuklarını çizelim 
plt.bar("Gymnastics", mens_gymnastics["Height"].mean(), yerr=mens_gymnastics["Height"].std())

plt.ylabel("Boy uzunlukları (cm)")

plt.show()
```

```python
# "Spor" sütununu seçip bunu sports_column adındaki değişkene atalım
sports_column = summer_2016["Sport"]

# "Spor" sütununun benzersiz/eşsiz değerlerini bulalım ve bunu sports adındaki değişkene atalım
sports = sports_column.unique()

print(sports)
```

```python
fig, ax = plt.subplots()

special_sports = ["Judo", "Rowing", "Taekwondo", "Handball", "Fencing", "Archery", "Equestrianism", "Boxing"]

# Fransa milli takımının farklı spor branşlarında yarışan sporcularının ağırlıklarının karşılaştırmasını for ile yapalım
for sport in special_sports:
  # Yalnızca o spor için satırları filtreleyelim
  sport_df = summer_2016[(summer_2016["Sport"] == sport) & (summer_2016["Team"] == "France")]
  # standart sapma (std) "yerr" hata çubuğuyla "Ağırlık" ortalaması için bir sütun grafiği oluşturalım
  ax.bar(sport, sport_df["Weight"].mean(), yerr=sport_df["Weight"].std())

plt.title("Fransa Milli Takımının Farklı Branştaki Oyuncularının Ağırlıklarının Karşılaştırılması")
plt.ylabel("Ağırlık")
ax.set_xticklabels(sports, rotation=90)

# Oluşturduğumuz şekli dosyaya kaydedelim
fig.savefig("sports_weights.png")
plt.show()
```
